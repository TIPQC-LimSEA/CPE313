```python
def changeFunc(sampleDict, C):
    sampleDict[1] = 1

    if C < 0:
        print("Invalid Amount!")
        return
    try:
        return sampleDict[C]
    except KeyError:
        if C >= 10:
            if C-10 == 0:
                print("10")
                return
                
            sampleDict[C] = (changeFunc(sampleDict, C-10))
            print("10")
            return sampleDict[C]
        if C >= 5:
            if C-5 == 0:
                print("5")
                return
            
            sampleDict[C] = (changeFunc(sampleDict, C-5))
            print("5")
            return sampleDict[C]
            
        if C < 5:
            if C-1 == 0:
                print("1")
                return
                
            sampleDict[C] = (changeFunc(sampleDict, C-1))
            print("1")
            return sampleDict[C]

testDict = {}
moneyVal = -10
print("Money:")
print(moneyVal)
print("Coins:")
changeFunc(testDict, moneyVal)
```
