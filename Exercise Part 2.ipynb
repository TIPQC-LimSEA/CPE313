```python
Read in the meteorite data from the Meteorite_Landings.csv file, rename the mass(g) column to mass, and drop all the latitude
and longitude columns. Sort the result by mass in descending order
```

```python
import pandas as pd

data = pd.read_csv("Meteorite_Landings.csv")

# Rename the mass (g) column to mass
data.rename(columns = {'mass(g)': 'mass'})

# drop all the latitude & longitude columns
data.drop()

# Sort the result by mass in descending order
data.sort_values(['mass', 'fall'], ascending = False)
```
